### Проверка расчета rougher.output.recovery

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке (test) для признака `rougher.output.recovery`. Найдем MAE между расчётами и фактическим значением признака. Для этого обратимся к описанию технологического процесса:

![TP](https://pictures.s3.yandex.net/resources/viruchka_1576238830.jpg)

**Технологический процесс:**
- `Rougher feed` — исходное сырье
- `Rougher additions` (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- `Xanthate` — ксантогенат (промотер, или активатор флотации);
- `Sulphate` — сульфат (на данном производстве сульфид натрия);
- `Depressant` — депрессант (силикат натрия).
- `Rougher process` (англ. «грубый процесс») — флотация
- `Rougher tails` — отвальные хвосты
- `Float banks` — флотационная установка
- `Cleaner process` — очистка
- `Rougher Au` — черновой концентрат золота
- `Final Au` — финальный концентрат золота

**Параметры этапов:**
- `air amount` — объём воздуха
- `fluid levels` — уровень жидкости
- `feed size` — размер гранул сырья
- `feed rate` — скорость подачи

**Наименование признаков:**
- `этап`.`тип_параметра`.`название_параметра`

**Возможные значения для блока `этап`:**
- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики

**Возможные значения для блока `тип_параметра`:**
- `input` — параметры сырья
- `output` — параметры продукта
- `state` — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики

Эффективность обогащения рассчитывается по формуле

![Recover](https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg)

где:
- `C` — доля золота в концентрате после флотации/очистки (**concentrate**);
- `F` — доля золота в сырье/концентрате до флотации/очистки (**feed**);
- `T` — доля золота в отвальных хвостах после флотации/очистки (**tail**).

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

Параметрам `C`, `F`, `T` соотвествуют следующие столбцы:
- `rougher.output.concentrate_au` 
- `rougher.input.feed_au`
- `rougher.output.tail_au`


```python
#запишем в виде функции для строки
def recovery(row):
    c = row['rougher.output.concentrate_au']
    f = row['rougher.input.feed_au']
    t = row['rougher.output.tail_au']
    return (c * (f - t)) / (f * (c - t)) * 100
```

Создадим урезанный датафрейм для простоты отображения:

```python
train_recovery = train[['rougher.output.concentrate_au', 
                            'rougher.input.feed_au', 
                            'rougher.output.tail_au', 
                            'rougher.output.recovery']]
```

Применим функцию построчно:

```python
train_recovery['recovery'] = train_recovery.apply(recovery, axis=1)
train_recovery
```

Вычислим метрику MAE:

```python
mae = mean_absolute_error(train_recovery['rougher.output.recovery'],
                    train_recovery['recovery'])
print(f'MAE вычисления эффективности обогащения {mae}')
#близко ли к 0
np.isclose(mae,0)
```

**Вывод**
- Метрика MAE близка к 0, что говорит о верном вычислении `rougher.output.recovery` в обучающей выборке

### Функция расчета sMAPE

Нужно спрогнозировать сразу две величины:
- эффективность обогащения чернового концентрата `rougher.output.recovery`
- эффективность обогащения финального концентрата `final.output.recovery`

![SMAPE](https://pictures.s3.yandex.net/resources/smape_1576238825.jpg)

- yi - target - Значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество
- y^i - prediction - Значение предсказания для объекта с порядковым номером i, например, в тестовой выборке

Итоговая метрика складывается из двух величин:

![SMAPE_final](https://pictures.s3.yandex.net/resources/_smape_1576238814.jpg)

Таким образом, у нас в фунции sMAPE будет на входе значения для `rougher.output.recovery` и `final.output.recovery` фактические и предсказанные

```python
def smape(target, predictions):
    numerator = np.abs(target - predictions)
    denominator = (np.abs(target) + np.abs(predictions))/2
    smape = sum(((numerator/denominator)*100)) / len(predictions)
    return smape

def smape_full(target, predictions):
    numerator = np.abs(target - predictions)
    denominator = (np.abs(target) + np.abs(predictions))/2
    smape = sum(((numerator/denominator)*100)) / len(predictions)
    return 0.25*smape[0] + 0.75*smape[1]
```

Судя по методу `describe` и гистограммам у большого количества параметров есть выбросы относительно визуально нормального распределения.
Для улучшения нормирования возьмем данные из 99.8% межквантильного интервала.

```python
def search_outliers(col, q1, q2):
    q_min = col.quantile(q1)
    q_max = col.quantile(q2)
    return col[(col > q_min) & (col < q_max)].index

q1 = 0.001
q2 = 0.999

for col in features_train_rougher.columns:
    index_rougher = search_outliers(features_train_rougher[col], q1, q2)
    features_train_rougher = features_train_rougher.loc[index_rougher]
    target_train_rougher = target_train_rougher.loc[index_rougher]

for col in features_train_final.columns:
    index_final = search_outliers(features_train_final[col], q1, q2)
    features_train_final = features_train_final.loc[index_final]
    target_train_final = target_train_final.loc[index_final]

print('Окончательные размеры выборок')   
print('Фичи и таргет rougher',features_train_rougher.shape, target_train_rougher.shape)   
print('Фичи и таргет final',features_train_final.shape, target_train_final.shape)   
```
