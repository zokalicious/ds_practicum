# Сборный проект "интернет-магазин «Стримчик»"

Работаем в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

У нас в наличии вами данные до 2016 года. На дворе сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Нужно отработать принцип работы с данными.

В наборе данных используется аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

## Подготовка данных

### Приведение названий столбцов к змеиному регистру

А также укоротим `Year_of_Release` для упрощения обращения:

```python
games.columns = games.columns.str.lower()
games = games.rename(columns={'year_of_release': 'release_year'})
games
```

```python
games.info()
```

### Преобразование типов данных

```python
games[games['user_score'] == 'tbd']
```

Исходя из общей информации необходимо исправить типы для столбцов `release_year` на *Int64* и `user_score` на *Int64*

В столбце `user_score` встречаются значения **tbd** (to be determined), что означает, что игра еще не получила достаточно отзывов от игроков или отзывы не опубликованы. Например, на портале metacritic отзывы пользователей начинают приниматься не ранее, чем через 3 дня после релиза игры. Заменим эти значения на NaN.

Дополнительно домножим столбец `user_score` на 10 для удобства сравнения с `critic_score` при необходимости и выставим тип *Int64*:

```python
games['release_year'] = games['release_year'].astype('Int64')
games['critic_score'] = games['critic_score'].astype('Int64')
games['user_score'] = pd.to_numeric(games['user_score'], errors='coerce')
games['user_score'] = 10*games['user_score']
games['user_score'] = games['user_score'].astype('Int64')
```

Проверим типы столбцов:

```python
games.info()
```

И выведем статистику по числовым столбцам:

```python
games.describe()
```

### Обработка пропущенных значений и проверка дубликатов

Подсчитаем количество пропущенных значений по столбцам:

```python
games.isna().sum()
```

```python
games[games['name'].isna()]
```

Не представляется возможным восстановить ни название игры, ни жанр, тем более релиз старый (1993 год). Удалим данные строки:

```python
games = games[~games['name'].isna() | ~games['genre'].isna()]
```

Пропуски в столбцах `name` и `genre` удалены

В столбцах `release_year`, `critic_score`, `user_score` пропуски заполнять не будем, так как это может повлиять на результат проверки статистических гипотез

Проверим датафрейм на предмет наличия полных явных дубликатов:

```python
games.duplicated().sum()
```

Явных дубликатов нет

Взглянем на столбец `rating`:

```python
list(games['rating'].unique())
```

Судя по системе **ESRB** играм присваиваются следующие категории:
- `EC` - early childhood - для детей младшего возраста
- `E` - everyone - для всех возрастов
- `E10+` - то же, что и `ET` - everyone 10 and older - для всех от 10 лет и старше
- `T` - teen - для подростков от 13 лет
- `M` - mature - для взлослых от 17 лет
- `AO` - adult only - только для врослых от 18 лет
- `RP` - rating pending - рейтинг ожидается (пока не определен)

Стоит отметить, что в датасете есть устаревшее значение `K-A` (kids-to-adults), которое теперь называется `E`. Соотвественно нужно внести правки в столбец с рейтингом. Также заменим `E10+` на более удобное `ET`.

```python
games.loc[games['rating'] == 'K-A', 'rating'] = 'E'
games.loc[games['rating'] == 'E10+', 'rating'] = 'ET'
```

*Смог найти датасет с рейтингами игр, который потенциально позволит заполнить пропуски в столбце `rating` (с kaggle)*:

```python
# подготовка списков с рейтингами
esrb_base = pd.read_csv('datasets/video_games_esrb_rating.csv')
esrb_base = esrb_base.loc[:, ['title', 'esrb_rating']]
esrb_name = list(esrb_base['title'])
esrb_rating = list(esrb_base['esrb_rating'])
```

```python
# список уникальных возрастных цензов
ratings = list(games['rating'].unique())
# без nan
del ratings[1]
ratings
```

Проверим соответствие по именам игр и заполним пропущенные значения:

```python
games_name = list(games['name'])
games_esrb = list(games['rating'])
count = 0
for index, game in enumerate(games_name):
    if games_esrb[index] not in ratings:
        try:
            games_esrb[index] = esrb_rating[esrb_name.index(game)]
            print(
                f'Найден рейтинг для строки номер {index}: {esrb_rating[esrb_name.index(game)]}')
            count += 1
        except:
            pass
print(f'Восстановлены значения возрастного ценза для {count} тайтлов')
games['rating'] = games_esrb
```

Проверим исправление в строке 72:

```python
games.loc[72, :]
```

### Продажи по платформам

```python
platform_pivot = games.pivot_table(index='platform', values={
                                   'total_sales'}, aggfunc='sum')
platform_pivot.reset_index(inplace=True)
platform_pivot.columns = ['platform', 'total_sales_sum']
platform_pivot = platform_pivot.sort_values('total_sales_sum', ascending=False)
plt.figure(figsize=(10, 4))
sns.barplot(data=platform_pivot, x='platform', y='total_sales_sum')
plt.title('Общее число релизов игр по годам')
plt.xticks(rotation=90)
plt.xlabel('Платформа')
plt.ylabel('Общее число релизов')
```

Выведем таблицу с продажами на платформах за все время. На конец 2016 года актуальными игровыми платформами являются PS4, XOne, PC, 3DS постараемся их захватить:

```python
platform_pivot.head(15)
```

Вычислим среднее время существования каждой платформы на рынке:

```python
span_pivot = games.pivot_table(index='platform', values={
                               'release_year'}, aggfunc={'max', 'min'})
span_pivot = span_pivot['release_year']
span_pivot['span'] = span_pivot['max'] - span_pivot['min']
span_pivot.head(10)
```

```python
span_pivot['span'].mean()
```

**Вывод**
- Средний срок жизни новой игровой платформы составляет 7.61 лет. Отдельно отметим, что игры на персональном компьютере всегда являются актуальным рынком, так как компьютерное железо можно апгрейдить и менять;
- Актуальным периодом будем считать такой, который бы охватил все наиболее популярные игровые платформы на момент 2016 года. Учитывая, что игровые платформы (кроме ПК) держатся на рынке, в среднем около 8 лет, то возьмем данные не старше 2-3 лет для анализа (рынок динамичный)

### Продажи в актуальный период

Возьмем в работу данные за соответствующий актуальный период, отбросим более старые данные:

```python
games_actual = games[games['release_year'] >= 2014]
```

```python
games_actual.describe()
```

```python
platform_pivot_actual = games_actual.pivot_table(
    index='platform', values={'total_sales'}, aggfunc='sum')
platform_pivot_actual.sort_values('total_sales', ascending=False)
platform_pivot_actual.reset_index(inplace=True)
platform_pivot_actual.columns = ['platform', 'total_sales_sum']
platform_pivot_actual = platform_pivot_actual.sort_values(
    'total_sales_sum', ascending=False)
plt.figure(figsize=(10, 4))
sns.barplot(data=platform_pivot_actual, x='platform', y='total_sales_sum')
plt.title('Число релизов игр по каждой платформе в актуальный период')
plt.xlabel('Платформа')
plt.ylabel('Общее число релизов')
```

### Выбор потенциально прибыльных платформ

Повторно построим сводную таблицу и гистограмму по актуальном периоду времени:

```python
games_actual_by_year = games_actual.pivot_table(
    index=['platform', 'release_year'],
    values='total_sales',
    aggfunc='sum')
games_actual_by_year = games_actual_by_year.sort_values(
    'total_sales', ascending=False).reset_index()
plt.figure(figsize=(14, 6))
ax = sns.barplot(x='release_year',
                 y='total_sales',
                 hue="platform",
                 data=games_actual_by_year)
plt.title('Общее число продаж игр по платформам в актуальный период')
plt.ylim(0, 140)
for i in ax.containers:
    ax.bar_label(i,)
ax.legend(bbox_to_anchor=(0.9, 1), loc='upper left')
plt.show()
```

```python
platform_pivot_actual.sort_values('total_sales_sum', ascending=False).head(10)
```

**Вывод**
- Стоит отметить, что 2016 год, в целом, выглядит неудачным для индустрии: продажи упали на всех платформах. Потенциально прибыльными платформами для релиза игр на 2017 год являются XOne, PS4, с некоторым отставанием следуют 3DS, WiiU;

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

- H_0: Средний пользовательский рейтинг `user_score` пользователей XOne = Средний пользовательский рейтинг `user_score` пользователей PC
- H_1: Средний пользовательский рейтинг `user_score` пользователей XOne ≠ Средний пользовательский рейтинг `user_score` пользователей PC
- alpha = 0.05

Для наглядности отобразим средние значения выборок:

```python
games_actual[games_actual['platform'] == 'XOne']['user_score'].mean()
```

```python
games_actual[games_actual['platform'] == 'PC']['user_score'].mean()
```

```python
xone_user_score = games_actual[(games_actual['platform'] == 'XOne') &
                               (~games_actual['user_score'].isna())].user_score.astype(float)
pc_user_score = games_actual[(games_actual['platform'] == 'PC') &
                             (~games_actual['user_score'].isna())].user_score.astype(float)
results = st.ttest_ind(xone_user_score, pc_user_score, equal_var=False)

alpha = 0.05

print(f'P-value гипотезы: {results.pvalue}')

# проверяем равенство средних
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу ')
```

**Вывод**
- Средние оценки игр на XOne за актуальный период 65.6%, а на PC - 62.9%
- Принимаем гипотезу H_0: средние пользовательские рейтинги игр на двух платформах равны

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

- H_0: Средний пользовательский рейтинг `user_score` жанра Action = Средний пользовательский рейтинг `user_score` жанра Sports
- H_1: Средний пользовательский рейтинг `user_score` жанра Action ≠ Средний пользовательский рейтинг `user_score` жанра Sports
- alpha = 0.05

Для наглядности отобразим средние значения выборок:

```python
games_actual[games_actual['genre'] == 'Action']['user_score'].mean()
```

```python
games_actual[games_actual['genre'] == 'Sports']['user_score'].mean()
```

```python
action_user_score = games_actual[(games_actual['genre'] == 'Action') &
                                 (~games_actual['user_score'].isna())].user_score.astype(float)
sports_user_score = games_actual[(games_actual['genre'] == 'Sports') &
                                 (~games_actual['user_score'].isna())].user_score.astype(float)
results = st.ttest_ind(action_user_score, sports_user_score, equal_var=False)

alpha = 0.05

print(f'P-value гипотезы: {results.pvalue}')

# проверяем неравенство средних
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу ')
```

**Вывод**
- Средние оценки игр жанра Action за актуальный период 67.6%, а Sports - 52.25%
- Применив ttest, принимаем гипотезу H_1: средние пользовательские рейтинги двух жанров не равны
