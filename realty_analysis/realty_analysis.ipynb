#### Работа с явными и неявными дубликатами

Проверяем наличие явных дубликатов:

```python
data.duplicated().sum()
```

Рассмотрим неявные дубликаты столбца `locality_name`:

```python
data['locality_name'].unique()
```

Изменим описание объектов: приведем все в нижний регистр и заменим ё на е, предварительно создав список с "короткими" названиями объектов:

```python
def prefix_remove(locality):
    locality = str(locality)
    for c in locality:
        if c.isupper():
            return locality[locality.index(c):len(locality)]


locality_name_short = list(data['locality_name'].apply(prefix_remove).unique())
locality_name_short
```

Посчитаем число вхождений из столбца коротких названий в полные названия для определения неявных дубликатов, где есть ошибка типа:

```python
for name in locality_name_short:
    st = ''
    count = 0
    for long_name in list(data['locality_name'].unique()):
        if str(name) in str(long_name):
            st += str(long_name)+' '
            count += 1
    if count > 1:
        print(st)
```

Сверимся с картой: очевидно существуют в ЛО деревни, села и поселки с одинаковыми именами. А вот наличие в списке, например **"поселок городского типа рощино"** и  **"городской поселок рощино"** - это явная ошибка. Таким образом, нужно переименовать тип всех объектов, где содержится "поселок" в префиксе на просто "поселок":

```python
def type_change(locality):
    locality = str(locality)
    if 'поселок' in locality or 'посёлок' in locality:
        for c in locality:
            if c.isupper():
                return 'поселок '+locality[locality.index(c):len(locality)]
    else:
        return locality


data['locality_name'] = data['locality_name'].apply(type_change)
data['locality_name'].unique()
```

```python
def naming_change(locality):
    locality = str(locality)
    locality_out = ''
    for c in locality:
        if c == 'ё':
            locality_out += 'е'
        else:
            locality_out += c.lower()
    return locality_out


data['locality_name'] = data['locality_name'].apply(naming_change)
data['locality_name'].unique()
```

Из законодательства:
*Для изолированной кухни минимальный размер — 8 кв. м.. Это требование не зависит от числа комнат в квартире.*

*В старых домах, которые проектировались до появления Свода правил, кухни были изначально меньше – 6, 5, или даже 4 кв. м. При ремонте в них действует одно правило: кухню нельзя уменьшать. А вот если мы проектируем новую изолированную кухню, тогда для согласования в Мосжилинспекции ее размер должен быть не менее 8. кв. м.*

Тут либо ошибка, либо нарушение планировки, допустим наличие кухонь не менее 4 кв.м.:

```python
data.loc[data['kitchen_area'] < 4, ['total_area', 'kitchen_area',
                                    'living_area', 'rooms', 'studio', 'is_apartment']]
```

Можно предположить, что это студии или квартиры с кухонными "уголками", а не апартаменты, соотвественно изменить для объектов с `kitchen_area` <  4 площадь на NaN, а `studio` и `is_apartment` на противоположные:

```python
data.loc[(data['kitchen_area'] < 4), 'studio'] = True
data.loc[(data['kitchen_area'] < 4), 'is_apartment'] = False
data.loc[(data['kitchen_area'] < 4), 'kitchen_area'] = np.nan
```

*Согласно действующим СНиПам, минимальная площадь комнаты должна быть не менее 8 квадратных метров (п. 5.3 СП 5.413330.2011), при этом её ширина не должна быть меньше 2,25 метра (п 6.1.9 СП 31-107-2004) При несоблюдении хотя бы одного из этих условий, считать комнату жилой нельзя.*

Найдем объекты, где размер жилой площади меньше 8 кв.м.:


```python
data.loc[data['living_area'] < 8, ['total_area', 'kitchen_area',
                                   'living_area', 'rooms', 'studio', 'is_apartment']]
```

Внесем правки:

```python
data.loc[data['studio'], 'is_apartment'] = False
# для студий вероятно перепутаны площадь жилая и площадь кухни
data.loc[(data['living_area'] < 8) & data['studio'],
         'living_area'] = data['kitchen_area']
# для квартир снова вычтем из общей площади площадь кухни для оценки верхней границы жилой площади
data.loc[(data['living_area'] < 8) & data['is_apartment'],
         'living_area'] = data['total_area'] - data['kitchen_area']
```

Проверим, есть ли квартиры, где совпадает статус студия-квартира:

```python
len(data[(data['studio'] & data['is_apartment'])])
```

```python
data.loc[(~data['studio']) & (~data['is_apartment']), ['total_area',
                                                       'kitchen_area', 'living_area', 'rooms', 'studio', 'is_apartment']]
```

Предположим, что это все же квартиры:

```python
data.loc[(~data['studio']) & (~data['is_apartment']), 'is_apartment'] = True
```

Типы квартир исправлены.

Проверим, не превышает ли `total_area` сумму площадей кухни и жилой:

```python
data.loc[data['living_area'] + data['kitchen_area'] > data['total_area'],
         ['total_area', 'kitchen_area', 'living_area', 'rooms', 'studio', 'is_apartment']]
```

Для таких объектов увеличим размер общей площади хотя бы до суммы жилой и кухни:

```python
data.loc[data['living_area'] + data['kitchen_area'] > data['total_area'], 'total_area'] \
    = data['living_area'] + data['kitchen_area']
```

#### Зависимость стоимости  квартиры от расстояния до центра города

Теперь выделим квартиры в Санкт-Петербурге с помощью столбца `locality_name` и вычислим среднюю цену каждого километра от центра по всем объявлениям в городе. Оформим в табличном виде:

```python
distance_data = data.loc[data['locality_name'] ==
                         'санкт-петербург', ['last_price', 'city_centers_nearest_km']]
distance_data
```

Оценим максимальное и минимальное расстояние от центра:

```python
print(
    f'Максимальное расстояние до центра: {distance_data["city_centers_nearest_km"].max()}км')
print(
    f'Минимальное расстояние до центра: {distance_data["city_centers_nearest_km"].min()}км')
```

Сделаем цикл по каждому километру с рассчетом средней стоимости квартиры в заданном радиусе и построим гистограмму::

```python
mean_km_price = []
for ind in range(30):
    mean_km_price.append(distance_data.query(
        '@ind <= city_centers_nearest_km < @ind+1')['last_price'].mean())
mean_km_price = {'average_price': mean_km_price}
mean_km_price = pd.DataFrame(mean_km_price)
mean_km_price
```

```python
mean_km_price.plot(y='average_price', style='o', xlabel='Расстояние в км от центра',
                   ylabel='Средняя цена', title='Зависимость средней цены жилья от расстояния от центра города в Санкт-Петербурге',
                   figsize=(15, 4))
```

**Вывод. Есть обратная зависимость цены объекта от расстояния до центра города, что логично. Стоит отметить наличие локального максимума в районе 5км радиуса относительно общего тренда, что может объясняться фондом новостроек с более высокой ценой на данном расстоянии от центра**
