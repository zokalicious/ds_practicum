### Разбиение выборки на обучающую, валидационную и тестовую

Воспользуемся библиотекой `fast_ml` для разбиения выборки на 3 части в отношении 3:1:1, все операции с использованием генератора псевдослучайных чисел будем использовать с фиксированным seed=555:

```python
features_train, target_train, features_valid, target_valid, features_test, target_test =\
    train_valid_test_split(data, target='exited',
                           train_size=0.6,
                           valid_size=0.2,
                           test_size=0.2,
                           random_state=12345)
features_train.shape
```

### OHE-кодирование категориальных признаков

Категориальные параметры `geography` и `gender` необходимо трансформировать в dummy-признаки методом OHE (удаляем первый столбец (для избежания дамми-ловушки):

```python
#категориальные признаки для OHE
ohe_features = data.select_dtypes(include='object').columns.to_list()
ohe_features
```

Подготовка кодировщика:

```python
encoder_ohe = OneHotEncoder(
                            drop='first',
                            handle_unknown='ignore',
                            sparse=False)
encoder_ohe.fit(features_train[ohe_features])
```

```python
encoder_ohe.get_feature_names_out()
```

Кодировка признаков в обучающей выборке:

```python
features_train[
    encoder_ohe.get_feature_names_out()
] = encoder_ohe.transform(features_train[ohe_features])
features_train = features_train.drop(ohe_features, axis=1)
features_train.columns = features_train.columns.str.lower()
features_train.head()
```

Кодировка признаков в валидационной выборке:

```python
features_valid[
    encoder_ohe.get_feature_names_out()
] = encoder_ohe.transform(features_valid[ohe_features])
features_valid = features_valid.drop(ohe_features, axis=1)
features_valid.columns = features_valid.columns.str.lower()
features_valid.head()
```

Кодировка признаков в тестовой выборке:

```python
features_test[
    encoder_ohe.get_feature_names_out()
] = encoder_ohe.transform(features_test[ohe_features])
features_test = features_test.drop(ohe_features, axis=1)
features_test.columns = features_test.columns.str.lower()
features_test.head()
```

Определим численные параметры и проведем для них нормализацию через `scaler`:

### Нормализация данных

Выделим столбцы с численными данными и нормализуем их значения:

```python
numeric = ['credit_score', 'age', 'tenure',
           'balance', 'num_of_products', 'estimated_salary']
#numeric = ['credit_score', 'age',
#           'balance', 'num_of_products']

scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
```

### Выбор модели обучения

Для решения задачи будем использовать следующие классы моделей и выберем лучшую с перебором различных параметров:
- Дерево решений (DecisionTreeClassifier)
- Случайный лес (RandomForestClassifier)
- Логистическая регрессия (LogisticRegression)


### DecisionTreeClassifier

Нужно перебрать в алгоритме обучения разные варианты глубины дерева, ограничимся максимальным значением `total_models` = 40:

```python
total_models = 40
f1_list = []
depth_list = [i for i in range(1, total_models+1)]
threshold_list = [0.59 for i in range(1, total_models+1)]
best_result = 0
for depth in range(1, total_models+1):
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model.fit(features_train, target_train)
    predictions_valid = model.predict(features_valid)
    f1 = f1_score(target_valid, predictions_valid)
    f1_list.append(f1)
    probabilities_valid = model.predict_proba(features_valid)
    auc_roc = roc_auc_score(target_valid, probabilities_valid[:, 1])
    if f1 > best_result:
        best_model_tree = model
        best_result = f1
        best_auc_roc = auc_roc
print(f'Модель с наивысшей F1: {best_model_tree}')
print(f'Метрика F1 модели {best_result}')
print(f'Метрика AUC-ROC модели {best_auc_roc}')
```

```python
acc_depth = pd.DataFrame(
    {'depth': depth_list, 'f1': f1_list, 'f1_threshold': threshold_list})
ax = acc_depth.plot(x="depth", y="f1",
                    marker="o",
                    figsize=(10, 4),
                    ylim=(0.5, 0.85))
acc_depth.plot(x="depth", y="f1_threshold",
               title='Зависимость метрики F1 от глубины дерева DecisionTreeClassifier',
               xlabel='Глубина дерева',
               ylabel='Метрика F1',
               ax=ax)
plt.show()
```

**Вывод**

Самая высокая метрика `F1` = 0.726301 для глубины дерева `max_depth` = 9

Данная модель показала результат `AUC-ROC` = 0.924128

### RandomForestClassifier

Нужно перебрать параметры n_estimators и max_depth:

```python
#зададим глобальные параметры для обучения
max_est = 40
max_depth = 20
```

```python
%%time
f1_list = []
estim_list = []
depth_list = []
threshold_list = []
best_result_forest = 0
for est in tqdm(range(1, max_est+1)):
    for depth in range(1, max_depth+1, 2):
        model = RandomForestClassifier(
            random_state=12345, n_estimators=est, max_depth=depth, n_jobs=-1)
        model.fit(features_train, target_train)
        predictions_valid = model.predict(features_valid)
        f1 = f1_score(target_valid, predictions_valid)
        f1_list.append(f1)
        estim_list.append(est)
        depth_list.append(depth)
        threshold_list.append(0.59)
        probabilities_valid = model.predict_proba(features_valid)
        auc_roc = roc_auc_score(target_valid, probabilities_valid[:, 1])
        if f1 > best_result_forest:
            best_model_forest = model
            best_result_forest = f1
            best_auc_roc_forest = auc_roc
print(f'Модель с наивысшей F1: {best_model_forest}')
print(f'F1 модели {best_result_forest}')
print(f'Метрика AUC-ROC модели {best_auc_roc_forest}')
```

```python
acc_depth_estim = pd.DataFrame({'estimators_number': estim_list, 'depth': depth_list,
                               'f1': f1_list, 'f1_threshold': threshold_list})
```

```python
plt.figure(figsize=(10, 4))
ax = sns.scatterplot(data=acc_depth_estim, x='estimators_number',
                     y='f1', hue='depth', palette='Spectral')
acc_depth_estim.plot(x="estimators_number", y="f1_threshold",
                     title='Зависимость метрики f1 от параметра n_estimators и max_depth метода RandomForestClassifier',
                     xlabel='Число estimators',
                     ylabel='Метрика f1',
                     color='Orange',
                     ylim=(0.45, 0.85),
                     ax=ax)
ax.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

**Вывод**

- Модель с наивысшей F1: RandomForestClassifier(max_depth=16, n_estimators=33, random_state=12345)
- F1 модели 0.757536
- Метрика AUC-ROC модели 0.941816

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(projection='3d')
surf =ax.plot_trisurf(acc_depth_estim['estimators_number'], acc_depth_estim['depth'], acc_depth_estim['f1'],
                      cmap=cm.Spectral,
                      linewidth=0.1)
fig.colorbar( surf, shrink=0.5, aspect=5)
ax.view_init(25, 250)
plt.xlabel('estimators')
plt.ylabel('depth')
plt.title('f1')
plt.show()
```

### LogisticRegression

Для логистической регрессии основным изменяемым гиперпараметром является число итераций iter. Проведем анализ по аналогии с пунктами 2.4 и 2.5. Число итераций будем брать кратным 50.

### Upsampling меньшего класса

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)

    return features_upsampled, target_upsampled
```

```python
features_upsampled, target_upsampled = upsample(
    features_train, target_train, 4)
features_upsampled.shape
```

```python
%%time
best_result_upsampled = 0
for est in tqdm(range(1, max_est+1)):
    for depth in range(1, max_depth+1):
        model_upsampled = RandomForestClassifier(
            random_state=12345, n_estimators=est, max_depth=depth, n_jobs=-1)
        model_upsampled.fit(features_upsampled, target_upsampled)
        predictions_valid = model_upsampled.predict(features_valid)
        f1 = f1_score(target_valid, predictions_valid)
        probabilities_valid = model_upsampled.predict_proba(features_valid)
        auc_roc = roc_auc_score(target_valid, probabilities_valid[:, 1])
        if f1 > best_result_upsampled:
            best_model_upsampled = model_upsampled
            best_result_upsampled = f1
            best_auc_roc_upsampled = auc_roc
print(f'Лучшая модель: {best_model_upsampled}')
print(f'F1 модели {best_result_upsampled}')
print(f'AUC-ROC модели {best_auc_roc_upsampled}')
```

**Вывод**

Метрика F1 также стала чуть выше, чем у модели без upsampling меньшего класса выборки

### Downsampling большего класса

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])

    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)

    return features_downsampled, target_downsampled
```

```python
features_downsampled, target_downsampled = downsample(
    features_train, target_train, 0.25)
features_downsampled.shape
```

```python
%%time
best_result_downsampled = 0
for est in tqdm(range(1, max_est+1)):
    for depth in range(1, max_depth+1):
        model_downsampled = RandomForestClassifier(
            random_state=12345, n_estimators=est, max_depth=depth, n_jobs=-1)
        model_downsampled.fit(features_downsampled, target_downsampled)
        predictions_valid = model_downsampled.predict(features_valid)
        f1 = f1_score(target_valid, predictions_valid)
        probabilities_valid = model_downsampled.predict_proba(features_valid)
        auc_roc = roc_auc_score(target_valid, probabilities_valid[:, 1])
        if f1 > best_result_downsampled:
            best_model_downsampled = model_downsampled
            best_result_downsampled = f1
            best_auc_roc_downsampled = auc_roc
print(f'Лучшая модель: {best_model_downsampled}')
print(f'F1 модели {best_result_downsampled}')
print(f'AUC-ROC модели {best_auc_roc_downsampled}')
```

**Вывод**

Downsampling большего класса в модели значительно снижает метрику F1
